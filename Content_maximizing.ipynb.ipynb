```python
content_df['Time of Day']=pd.to_datetime(content_df["Time of Day"],dayfirst=True)
content_df['Merged']=pd.to_datetime(content_df["Merged"],dayfirst=True)
content_df["Peak Viewership Time"]=pd.to_datetime(content_df["Peak Viewership Time"],dayfirst=True)
```

```python
content_df["viewership_bins"]=pd.cut(content_df["Viewership"],bins=5)
content_df["Shares_bins"]=pd.cut(content_df["Shares"],bins=5)
content_df["Comments_bins"]=pd.cut(content_df["Comments"],bins=5)
content_df["Likes_bins"]=pd.cut(content_df["Likes"],bins=5)
```

```python
content_df["viewership_bins"]=content_df["viewership_bins"].astype('str')
content_df["Shares_bins"]=content_df["Shares_bins"].astype('str')
content_df["Comments_bins"]=content_df["Comments_bins"].astype('str')
content_df["Likes_bins"]=content_df["Likes_bins"].astype('str')
```

```python
# What are the top 10 most viewed content pieces overall?
Top_content=content_df.groupby(by=["Content ID"]).agg({"Viewership":"sum"}).reset_index()
Top_content.sort_values(by=["Viewership"],ascending=[False],inplace=True)
Top_content.iloc[0:10,:].reset_index(drop=True)
```

```python
# What are the most-watched genres in each region?
genre_by_location_views=content_df.groupby(["Viewer Location","Genre"]).agg({"Viewership":"sum"}).reset_index()
genre_by_location_views.sort_values(by=["Viewership","Viewer Location"],ascending=[False,True],inplace=True)
genre_by_location_views=genre_by_location_views.reset_index(drop=True)
genre_by_location_views['genre_rank']=genre_by_location_views.groupby("Viewer Location")["Viewership"].rank(method='min',ascending=False)
genre_by_location_views.loc[genre_by_location_views["genre_rank"]==1.0,:]
```

```python
# Which genre has the highest average watch time?
viewership_by_genre=content_df.groupby(["Genre"]).agg({"Viewership":np.average}).reset_index()
viewership_by_genre=viewership_by_genre.sort_values(by=["Viewership"],ascending=[False]).reset_index(drop=True)
# viewership_by_genre.iloc[[0],:]
viewership_by_genre
```

```python
# What time of day has the highest viewer engagement?
peak_viewership_by_genre=content_df.groupby(by=["Peak Viewership Time"]).agg({"Viewership":"sum"}).reset_index()
peak_viewership_by_genre=peak_viewership_by_genre.sort_values(by=["Viewership"],ascending=[False]).reset_index(drop=True)
peak_viewership_by_genre.iloc[[0],:]
```

```python
# Which days of the week drive the most views?​
views_by_week_day=content_df.groupby(by=["Peak Viewership Day"]).agg({"Viewership":"sum"}).reset_index()
views_by_week_day.iloc[[0],:]
```

```python
# What is the average user rating for different content types (Movies vs Series)?
rating_by_content_type=content_df.groupby(by=["Content Type"]).agg({"User Ratings":np.average}).reset_index()
rating_by_content_type=rating_by_content_type.sort_values(by=["User Ratings"],ascending=[False])
rating_by_content_type.reset_index(drop=True)
```

```python
#How many likes, shares, and comments are generated per content type?
likes_by_content_type=content_df.groupby(by=["Content Type"]).agg({"Likes":"sum",
                                             "Shares":"sum",
                                             "Comments":"sum"}).reset_index()
likes_by_content_type=likes_by_content_type.sort_values(by=["Likes","Shares","Comments"],ascending=[False,False,False])
likes_by_content_type.reset_index(drop=True)
```
